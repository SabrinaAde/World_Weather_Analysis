```python
import random 
import timeit
import numpy as np
import pandas as pd
from citipy import citipy
import requests
from config import weahter_api_key
import json
```

```python

latitude = []
longitude = []

for each_lat in np.random.randint(-90,90,2000):
    latitude.append(each_lat)

for each_long in np.random.randint(-180,180,2000):
    longitude.append(each_long)

```

```python
#store all random longitudes and latitudes
latitude_and_Longitude = tuple(zip(latitude,longitude))
```

```python
#getting nearest city from citipy

# City= citipy.nearest_city(22.99, 120.21)

```

```python
#use citipy  module to find nearest city for all latitude and longitudes, faster than pulling up websites endpoint
cities = []
country = []

for lat,lon in latitude_and_Longitude:
    city = citipy.nearest_city(lat,lon)
    
    cityName = city.city_name
    
    cities.append(cityName)
    
    country.append(city.country_code)


```

```python
df_countries = pd.DataFrame({'Cities': cities,
'Country': country})

df_countries['Latitude']= ''
df_countries['Longitude']= ''
df_countries['Temperature (F)']= ''
df_countries['Humidity (%)']= ''
df_countries['Cloudiness (%)']= ''
df_countries['Wind Speed (mph)']= ''

#drop all duplicates in column Cities
df_countries = df_countries.drop_duplicates(subset=['Cities'], keep= 'first')


#loop through all rows and fill in values for blank columns in dataframes

for index, row in df_countries.iterrows():
    city_name= row['Cities']
    Country_id=row['Country']

#define api url from openweather map website
    api_url = "http://api.openweathermap.org/data/2.5/forecast" \
    "?q={},{}&units=IMPERIAL&mode=json&APPID={}".format(city_name, Country_id,weahter_api_key)

    #place api request from openweather map for all cities
country_info = requests.get(api_url).json()
    
    #retrieve key information from api call such as max temp, latitude, longitude etc

```

```python

# df_countries.to_csv("C:\\Users\\sabri\\Desktop\\Class Projects\\Resources\\country_dataframe.csv")
try:
        # df_countries.at[1,'Latitude']= country_info['coord']['city']['lat']
     df_countries.at['Latitude']=country_info['city']['coord']['lat']
except KeyError:
    df_countries.at['Latitude']= np.nan
    # #     df_countries.at(index, 'Longitude', np.nan)

try:
        # df_countries.at[1,'Latitude']= country_info['coord']['city']['lat']
     df_countries.at['Longitude']=country_info['city']['coord']['lon']
except KeyError:
    df_countries.at['Longitude']= np.nan
    # #     df_countries.at(index, 'Longitude', np.nan)

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weahter_api_key

#create an endpoint URL for a city
# city_url = url + "&q=" + "Boston"

# city_weather = requests.get(city_url)
#getting response 200 means that we succesfully connected and retrieved data from the server

#get text from the 'get' request
# city_weather.text

# # Create an endpoint URL for a city to test that the call gets a valid response 
# city_url = url + "&q=" + "Boston"
# city_weather = requests.get(city_url)
# if city_weather.status_code == 200:
#     print(f"City Weather found.")
# else:
#     print(f"City weather not found.")


import time

from datetime import datetime

# Create an empty list to hold the weather data.
city_data = []
# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1

#loop through all the cities in list in groups of 50

# for i in range(len(cities)):

#     if (i% 50==0 and i >= 50):
#         set_count += 1
#         record_count =1
#         #delay execution for 60 seconds
#         # time.sleep(60)
#     #create endpoint url with each city
#     city_url = api_url + "&q="  + cities[i]

#loop through all the cities in the list
for i, city in enumerate(cities):
    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
        # time.sleep(60)

    # Create endpoint URL for each city.
    city_url = url + "&q=" + city.replace(" ","+")

    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    # Add 1 to the record count.
    record_count += 1


#set try-except in the event api request stops prematurely (finds empty data set)
# Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                            "Lat": city_lat,
                            "Lng": city_lng,
                            "Max Temp": city_max_temp,
                            "Humidity": city_humidity,
                            "Cloudiness": city_clouds,
                            "Wind Speed": city_wind,
                            "Country": city_country,
                            "Date": city_date})


# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")


```

```python
#convert dictionary list of city weather data to pandas

city_data_df = pd.DataFrame(city_data)

new_column_order = ['City', 'Country', 'Date', 'Lat', 'Lng', 'Max Temp','Humidity', 'Cloudiness','Wind Speed']

city_data_df = city_data_df[new_column_order]

city_data_df.head(10)

# #create output file

output_data_file = 'Weather_Database\\Weather_Database.csv'


city_data_df.to_csv(output_data_file, index_label='City_ID')
```
